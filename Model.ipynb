```python
import pandas as pd
import requests
from io import BytesIO
from PIL import Image
import torch
from transformers import AutoTokenizer, AutoModel
import torchvision.transforms as T
from torchvision.transforms.functional import InterpolationMode
from concurrent.futures import ThreadPoolExecutor

# Constants
IMAGENET_MEAN = (0.485, 0.456, 0.406)
IMAGENET_STD = (0.229, 0.224, 0.225)

def build_transform(input_size):
    MEAN, STD = IMAGENET_MEAN, IMAGENET_STD
    transform = T.Compose([
        T.Lambda(lambda img: img.convert('RGB') if img.mode != 'RGB' else img),
        T.Resize((input_size, input_size), interpolation=InterpolationMode.BICUBIC),
        T.ToTensor(),
        T.Normalize(mean=MEAN, std=STD)
    ])
    return transform

# Download image from URL
def download_image(url):
    try:
        response = requests.get(url)
        response.raise_for_status()
        img = Image.open(BytesIO(response.content))
        return img.convert('RGB')
    except Exception as e:
        print(f"Error downloading image from {url}: {e}")
        return None

# Load image URLs and their associated entity values from a CSV file
def load_image_data(csv_path):
    try:
        df = pd.read_csv(csv_path, dtype=str)  # Ensure all data is read as strings
        return df[['index', 'image_link', 'entity_name']].to_dict('records')
    except Exception as e:
        print(f"Error reading CSV: {e}")
        return []

def process_batch(image_data, input_size=224, max_num=5):
    with ThreadPoolExecutor() as executor:
        processed_images = []
        valid_data = []
        futures = [executor.submit(download_image, data['image_link']) for data in image_data]
        for future, data in zip(futures, image_data):
            image = future.result()
            if image is not None:
                transform = build_transform(input_size)
                processed_image = transform(image)
                processed_images.append(processed_image)
                valid_data.append(data)

    # Check if any images were processed
    if not processed_images:
        return torch.empty(0), valid_data

    pixel_values = torch.stack(processed_images)
    return pixel_values, valid_data

def batch_infer_images(image_data, model, tokenizer, input_size=224, max_num=5):
    pixel_values, valid_data = process_batch(image_data, input_size=input_size, max_num=max_num)

    if pixel_values.numel() == 0:
        print("No valid images were processed.")
        return [], valid_data

    pixel_values = pixel_values.to(torch.bfloat16).cuda()
    num_images = len(valid_data)

    # Create entity-specific questions
    questions = [
        f"Analyse the image very carefully. Find the {data['entity_name']} of the item which is given in the image. Output the value with its unit only and nothing else. If you don't find any value, check again."
        for data in valid_data
    ]

    num_patches_list = [pixel_values.size(0) // num_images] * num_images  # Adjust num_patches_list based on batch size

    # Debugging
    print(f"Pixel values shape: {pixel_values.shape}")
    print(f"Num patches list: {num_patches_list}")
    print(f"Questions count: {len(questions)}")

    # Ensure lengths match
    if len(questions) != len(num_patches_list):
        print(f"Warning: Number of questions ({len(questions)}) does not match number of images ({num_images})")
        # Adjust num_patches_list to match the length of questions if necessary
        num_patches_list = [len(questions) // len(num_patches_list)] * len(num_patches_list)

    # Perform batch inference
    try:
        responses = model.batch_chat(
            tokenizer,
            pixel_values,
            num_patches_list=num_patches_list,
            questions=questions,
            generation_config={"max_new_tokens": 256, "do_sample": False}
        )
    except IndexError as e:
        print(f"Batch inference failed with IndexError: {e}. Processing images individually.")
        # Process each image individually if batch processing fails
        responses = []
        for idx in range(num_images):
            single_image_pixel_values = pixel_values[idx:idx+1]
            single_question = questions[idx]
            try:
                response = model.batch_chat(
                    tokenizer,
                    single_image_pixel_values,
                    num_patches_list=[single_image_pixel_values.size(0)],
                    questions=[single_question],
                    generation_config={"max_new_tokens": 256, "do_sample": False}
                )
                responses.extend(response)
            except Exception as e:
                print(f"Error processing image {idx} individually: {e}")
                responses.append("Error")

    # Debugging
    print(f"Responses count: {len(responses)}")

    return responses, valid_data

def save_results_to_csv(csv_output_path, results, valid_data):
    num_results = len(results)
    num_valid_data = len(valid_data)

    print(f"Number of results: {num_results}")
    print(f"Number of valid data entries: {num_valid_data}")

    if num_results != num_valid_data:
        print(f"Warning: Number of results does not match number of valid data entries.")

    num_entries = min(num_results, num_valid_data)

    output_data = {
        'index': [data['index'] for data in valid_data[:num_entries]],
        'entity_name': [data['entity_name'] for data in valid_data[:num_entries]],
        'model_output': results[:num_entries]
    }

    df = pd.DataFrame(output_data)
    df.to_csv(csv_output_path, mode='a', index=False, header=not pd.io.common.file_exists(csv_output_path))


# Main batch processing flow
csv_input_path = '/content/test_chunk_1.csv'  # Path to your input CSV file
csv_output_path = '/content/drive/MyDrive/AMAZON_ML/test_chunk1/output_chunk_1.csv'  # Path to your output CSV file

# Load the image data from the input CSV
image_data = load_image_data(csv_input_path)

# # Load your model and tokenizer (replace with actual model and tokenizer)
# model = AutoModel.from_pretrained('your-model')
# tokenizer = AutoTokenizer.from_pretrained('your-model')

batch_size = 5

# Process images in smaller batches and print results as they are processed
for i in range(0, len(image_data), batch_size):
    batch_data = image_data[i:i+batch_size]
    batch_results, valid_data = batch_infer_images(batch_data, model, tokenizer, input_size=448, max_num=5)

    # Save the batch results to an output CSV file
    save_results_to_csv(csv_output_path, batch_results, valid_data)

    # Print results as they are processed
    for result, data in zip(batch_results, valid_data):
        print(f"Entity: {data['entity_name']}\nAssistant: {result}")
```
